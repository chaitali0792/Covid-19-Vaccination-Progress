# Correlations

```python
df_CVN.columns
```

```python
# Creating a subset to exclude the columns Country, iso_code, Date, vaccines, source_name, source_website
sub_CVN = df_CVN[['Total_No_Of_vaccinations', 'Total_No_People_Vaccinated', 'Total_No_Of_People_Fully_Vaccinated', 
                  'daily_vaccinations_raw','daily_vaccinations', 
                  'total_vaccinations_per_hundred', 'Total_No_People_Vaccinated_Per_Hundred', 
                  'Total_No_OF_People_Fully_Vaccinated_Per_Hundred', 'daily_vaccinations_per_million'
                 ]]
```

```python
sub_CVN.head()
```

```python
# Create a correlation matrix using pandas
sub_CVN.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub_CVN.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(9, 9)) # figure size 
plt.matshow(sub_CVN.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub_CVN.shape[1]), sub_CVN.columns, fontsize=12, rotation=45) # x axis labels
plt.yticks(range(sub_CVN.shape[1]), sub_CVN.columns, fontsize=12) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

Create a correlation heatmap using seaborn:

```python
# Create a subplot with matplotlib.

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub_CVN.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

Correlation plots using matplotlib and seaborn library have been created to check relation between the variables.
The plot using matplotlib is quite difficult to understand as compared to seaborn. On other hand, plot using seaborn library is easy to read as it has coefficient values on it.
From heatmap, we can see that 'Total_No_Of_vaccinations','Total_No_People_Vaccinated' and 'Total_No_Of_People_Fully_Vaccinated' these columns have high value of coefficient with each other. For example, if we consider the total number of vaccination as a dependent variable and total number of vaccinated or fully vaccinated people as independent variables, then we have strong co-relationship between these variables. As the total number of vaccinated or full vaccinated people increases the total vaccinations number also increase.

Above scatterplots have been created to check whether relationship between the total number of vaccinations against the total number of vaccinated or fully vaccinated people is linear or not.
In the first scatterplot, you can see that there is strong positive relationship with trend line is steep, and there are many points closer to this line. 
From the second scatterplot, we still have a strong postive trend but points are less closer to the line.

# Categorical Plots:

```python
#Creating subset to only include any one country

df_south_Africa = df_CVN[df_CVN['Country'] == 'South Africa']
```

```python
df_south_Africa.head()
```

```python
# Using a histogram to visualize the distribution of the variables. 

sns.distplot(df_south_Africa['Total_No_Of_vaccinations'], bins = 5)
```

```python
df_south_Africa.loc[(df_south_Africa['Total_No_Of_vaccinations'] < 2000000),'Vaccination Category'] = 'Low vaccination'
```

```python
df_south_Africa.loc[(df_south_Africa['Total_No_Of_vaccinations'] >= 2000000) 
& (df_south_Africa['Total_No_Of_vaccinations'] < 5000000) , 'Vaccination Category'] = 'Middle vaccination'
```

```python
df_south_Africa.loc[df_south_Africa['Total_No_Of_vaccinations'] >= 5000000 , 'Vaccination Category'] = 'High vaccination'
```

```python
df_south_Africa['Vaccination Category'].value_counts(dropna = False)
```

```python
# Creating a categorical plot in seaborn using the vaccination categories created above
sns.set(style="ticks")
cat_plot = sns.catplot(x="Total_No_People_Vaccinated", y="Total_No_Of_vaccinations", hue="Vaccination Category", data=df_south_Africa)
```

I have created subset for South Africa as we have more number of records and I was not getting the proper histogram for the distribution of total number of vaccinations. With this subset, I created categories for Total number of vaccinations as Low, Medium and High.
From the above categorical plot, we can see that total number of vcaccinated people are mostly from low vaccination group and almost same number of vaccinated people are from middle and high vaccination group which is less as compared to low vaccination. That means most of the people have got their first dose and few people got their second doses.
